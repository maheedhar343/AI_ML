```python
import pydicom
import numpy as np
import matplotlib.pyplot as plt
def dicom_to_numpy(file_path):
    dcm_file = pydicom.dcmread("/content/1000.dcm")
    pixel_array = dcm_file.pixel_array
    return pixel_array

def display_image(pixel_array):
    plt.imshow(pixel_array, cmap='gray')  # Adjust cmap if needed
    plt.axis('off')
    plt.show()
file_path = "/content/1000.dcm"
pixel_array = dicom_to_numpy(file_path)
display_image(pixel_array)
```
