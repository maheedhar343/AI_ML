```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd
```

```python
zomato_cafe_data = "/content/reviews.csv"
```

```python
zomato_df = pd.read_csv(zomato_cafe_data,index_col = 'Index')
zomato_df.head()
```

```python
zomato_df.isnull().sum()
```

```python
desired_city = "ahmedabad"
res = zomato_df[zomato_df["City"]==desired_city]["Review"]
# Convert to Python List
res_l = res.tolist()
# Combine list of reviews into a single string
res_text = " ".join(res_l)
```

```python
n = 30
wc = WordCloud(width=1000,height=500,max_words=n,background_color="white",min_font_size=10).generate(res_text)
```

```python
# plt.figure(figsize=(10,5))
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
